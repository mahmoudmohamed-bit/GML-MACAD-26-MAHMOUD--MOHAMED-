## 4. Utility functions to reset the face dictionaries and transfer dictionaries by key

```python
def reset_dictionaries(shell):
    faces = Topology.Faces(shell)
    for i, f in enumerate(faces):
        d = Topology.Dictionary(f)
        keys = Dictionary.Keys(d)
        for key in keys:
            if not key == "face_id":
                d = Dictionary.RemoveKey(d, key)
        f = Topology.SetDictionary(f, d)

def transfer_dicts_by_key(topologies, selectors, key):
    dicts = {}
    for t in topologies:
        d = Topology.Dictionary(t)
        value = Dictionary.ValueAtKey(d, key, None)
        if value:
            dicts[str(value)] = t
    
    for s in selectors:
        d = Topology.Dictionary(s)
        value = Dictionary.ValueAtKey(d, key, None)
        if value:
            f = dicts[str(value)]
            f = Topology.SetDictionary(f, d)
```

## 7. MESH EDGE REMOVAL 

```python
from shapely.geometry import Polygon as ShapelyPolygon
from shapely.ops import unary_union

def shapely_to_face(polygon):
    outer = Wire.ByVertices([Vertex.ByCoordinates(x, y, 0.0) for x, y in list(polygon.exterior.coords)[:-1]], close=True)
    holes = [Wire.ByVertices([Vertex.ByCoordinates(x, y, 0.0) for x, y in list(i.coords)[:-1]], close=True) for i in polygon.interiors]
    return Face.ByWires(outer, holes) if holes else Face.ByWire(outer)

tri_faces = Topology.Faces(Topology.ByBREPPath(r"D:\MACAD\3rd-SM\GML\python\GML.26\Asset\resi floor_export_surface.brep"))
dissolved = unary_union([ShapelyPolygon([(v.X(), v.Y()) for v in Topology.Vertices(f)]) for f in tri_faces])

if dissolved.geom_type == "MultiPolygon":
    parts = [shapely_to_face(p) for p in dissolved.geoms if p.area > 1e-8]
    parts = [(Topology.RemoveCollinearEdges(f) or f) for f in parts if f is not None]
    gallery = Cluster.ByTopologies(parts) if len(parts) > 1 else parts[0]
else:
    raw = shapely_to_face(dissolved)
    gallery = Topology.RemoveCollinearEdges(raw) or raw
```

### d. Closeness Centrality/Integration
* Closeness centrality is a graph metric that quantifies how close a node is to all other nodes by taking the reciprocal of the sum of its shortest path distances to every other node in the network.
* In space syntax, closeness centrality corresponds to global integration, measuring how spatially accessible or topologically shallow a space is within a configuration, thereby indicating its potential for movement flow and encounter density.
